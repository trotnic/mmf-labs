```python
def recursive_search(sus: list, term: str, i: int):
    
    # 1 + 1 + 1 + 1
    if term and i < len(sus):
        # 1
        num = -1
        # 2 + 1
        if sus[i] == term:
            # 1
            num = i
        else:
            # 1
            if len(sus):
                recursive_search(sus,term, i+1)
        # 1
        if num == -1:            
            print("Такого ключа в массиве нет")
        else:
            print(f"Ключ {num} найден")
# O(11n) ~ O(n)
    
```
