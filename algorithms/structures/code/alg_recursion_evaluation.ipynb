```python
def recursive_search(sus: list, term: str, i: int):
    
    # 1 + 1 + 1 + 1
    if term and i < len(sus):
        # 1
        num = -1
        # 2 + 1
        if sus[i] == term:
            # 1
            num = i
        else:
            # 1
            if len(sus):
                recursive_search(sus,term, i+1)
        # 1
        if num == -1:            
            print("Такого ключа в массиве нет")
        else:
            print(f"Ключ {num} найден")
# O(11n) ~ O(n)
    
```

```python
def random_substring(string: str) -> str:
    seed(t.time())
    length = randint(2, 40)
    pos = randint(0,len(string))
    return string[pos:pos+abs(length)]
```

```python
def random_main_string(string: str) -> str:
    start = randint(0,len(string))
    finish = randint(start, len(string))
    return string[start:finish]
```

```python
def evaluate_results(repetitions: int, string) -> list:
    meta = []
    for _ in range(repetitions):
        main_string = random_main_string(string)
        sus = random_substring(string)
        start_time = t.time()
        recursive_search(main_string, sus, 0)
        finish_time = t.time()
        meta.append((len(main_string), finish_time - start_time))
    return meta
```

```python
data = evaluate_results(10,SUS)
```

```python
def build_chart(raw):
    chart_data = []

    for (n, time) in raw:
        chart_data.append(
            dict(
                size=n,
                evaluation=time
            )
        )
    fig = px.line(chart_data, x="size", y="evaluation")
    fig.show()
```

```python
build_chart(sorted(data, key=lambda x:x[0]))
```
