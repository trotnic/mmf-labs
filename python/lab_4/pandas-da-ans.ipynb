```python
# - Происхождение
# Люди какого происхождения чаще всего подвергаются преступлениям?
```

```python
df = data['Victim Descent'].value_counts().sort_values()
df.plot(x='Victim Descent', kind="barh", colormap="tab10", figsize=(6,6));
```

```python
# Место проишествия
# Отсортируйте районы, по количеству преступлений. Постройте график, показывающий самые безопасный и опасный районы.
```

```python
obj = data['Area Name'].value_counts()
obj.sort_values().plot(kind="barh", colormap='tab10', figsize=(6,6));
```

```python
# Преступления, пол и возраст
# Изучите распределение количества преступлений по возрасту. Какова тенденция? Люди какого возраста чаще всего подвергаются преступлениям? Есть ли локальные минимумы? Используйте типы графиков hist и density.

data['Victim Age'].plot.kde(figsize=(6,6));
```

```python
# Определите 10 самых распространённых преступлений в LA. Постройте график.
crimes = data[data['Area Name'].str.contains('LA')]['Crime Code Description'].value_counts().head(10)
crimes.sort_values().plot(kind="barh", figsize=(6,6), colormap="tab10");
```

```python
# Люди какого происхождения чаще всего страдают от преступлений в каждом из районов? 

df = data.groupby('Area Name')['Victim Descent'].value_counts().transform(lambda df: (df - df.min()) / (df.max() - df.min()))
df.unstack().plot(figsize=(20,20), kind='barh', linewidth=1, colormap='prism', linestyle=':')
```

```python
# Как различается вероятность женщин и мужчин стать жертвой в зависимости от возраста? Постройте визуализацию. В каком возрастном промежутке мужчины чаще становятся жетрвами преступлений?

df_agg = data.groupby('Victim Age')['Victim Sex'].value_counts()
df_agg.unstack().plot(figsize=(10,10), colormap="bwr");
```

```python
# От каких преступлений чаще старадют женщины, а от каких мужчины?

df = data.groupby('Victim Sex')['Crime Code Description'].value_counts().unstack()

females = df.loc['F'].sort_values(ascending=False).head(5)
males = df.loc['M'].sort_values(ascending=False).head(5)

df = pd.DataFrame({'Female': females, 'Male': males})
df.plot(kind='bar', colormap='tab10', figsize=(6,6));
```

```python
# В наборе данных имеется информация о Возрасте, Поле, и Происхождении каждой жертвы. Есть ли связь между этими признаками?
# Верно ли, что женщины чаще оказываются жертвами по сравнению с мужчинами?
import seaborn as sns
obj = data[['Victim Age', 'Victim Sex', 'Victim Descent']].value_counts()
sns.set(rc={'figure.figsize':(10,10)})
sns.set(style="darkgrid")
sns.countplot(data=obj.reset_index(), x='Victim Descent', hue='Victim Sex',palette='Set3');
```
